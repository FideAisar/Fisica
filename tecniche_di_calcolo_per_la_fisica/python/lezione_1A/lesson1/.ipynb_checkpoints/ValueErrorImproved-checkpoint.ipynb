### Come scrivere correttamente misure e incertezze (migliorato)

```python
def notScient(numero, potenza = -99) :

    if (potenza != -99) :
        return numero / (10 ** potenza) , potenza
    else:
        potenza = 0
        while (numero < 1.) :
            potenza -= 1 ;   numero *= 10. 
        while (numero >= 10.) :
            potenza += 1 ;   numero /= 10.
            
    return numero, potenza
```

```python

valore = 0.00000478
errore = 0.00000034
```

```python
divErr, potErr = notScient(errore)

if (divErr < 2.) : 
    potErr -= 1
    divErr, potErr = notScient(errore, potErr)
    
divVal, potErr = notScient(valore, potErr)    

# Evitare notazioni ridicole con potenze troppo basse
potenzaMinima = 2
if (abs(potErr) > potenzaMinima) :
    print (f'Valore +/- errore = ({divVal:.0f} +/- {divErr:.0f}) x 10^{potErr}')
elif (potErr >= 0) :
    divVal, divErr = round(divVal) * (10 ** potErr), round(divErr) * (10 ** potErr)
    print (f'Valore +/- errore = {divVal} +/- {divErr}')
else: 
    potErr = -potErr
    print (f'Valore +/- errore = {valore:.{potErr}f} +/- {errore:.{potErr}f}')
    
```

* `abs` e `round` sono funzioni matematiche *incorporate* in Python.
* Altre che non lo sono possono essere chiamate tramite il modulo `math` (lo vedremo poi).

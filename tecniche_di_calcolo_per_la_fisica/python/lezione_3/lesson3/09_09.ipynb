# 1) Tentativi

### Istruzione `try` 
* Un'istruzione `try` introduce una parte di codice che potrebbe potenzialmente portare degli errori in runtime.
* La suite di `try` viene eseguita se non produce errori, altrimenti viene eseguita la suite di `except`
* Ci può essere alla fine un'istruzione `finally` dopo ogni istruzione `except` oppure `else`, che viene sempre eseguita

### Esempio

```python
try:
    print('try suite with no exceptions raised')
except:
    print('this will not execute')
finally:  
    print('finally always executes')
    
```

```python
try:
    print('try suite that raises an exception')
    int('hello')
    print('this will not execute')
except ValueError:
    print('a ValueError occurred')
finally:  
    print('finally always executes')
    
```

### Combinare `with` con `try…except`  
* Esempio: in un codice di processamento di file _affidabile_, la lettura del file appare sempre in una `try` suite che contiene un'istruzione `with`, per garantire che la risorsa sia utilizzabile 

```python
open('gradez.txt')  # non-existent file
```

```python
try:
    with open('gradez.txt', 'r') as accounts:
        print(f'{"ID":<3}{"Name":<7}{"Grade"}')
        for record in accounts:  
            student_id, name, grade = record.split()
            print(f'{student_id:<3}{name:<7}{grade}')
except FileNotFoundError:
    print('The file name you specified does not exist')
    
```

# 2) Sollevare un'eccezione esplicitamente
* Alcune funzioni invece devono informare esplicitamente il chiamante del fatto che è avvenuto un errore
* L'istruzione **`raise`** solleva un'eccezione di questo tipo:
```python
raise ExceptionClassName
```
* Crea un oggetto della classe di eccezioni specificata
* Questo nome può essere seguito da una stringa tra parentesi che spiega in dettaglio l'errore che è avvenuto (stringa di inizializzazione dell'oggetto eccezione
* Il codice che solleva l'eccezione deve essere sempre posto **prima** dell'uso di qualsiasi istruzione per cui serve che l'errore non sia avv
* E' raccomandato usare uno dei [tipi di eccezione predefiniti](https://docs.python.org/3/library/exceptions.html)

# I comandi "magici" di Python: `%timeit` 
* Il **comando magico`%timeit`** misura il tempo **medio** per l'esecuzione dell'operazione
* Di default, `%timeit` esegue un'istruzione in un ciclo e gira il ciclo 7 volte
* Se non indicate il numero di loop, `%timeit` sceglie lui un valore appropriato basandosi sul tempo della prima esecuzione
* Alla fine, `%timeit` mostra il tempo di esecuzione _medio_ e la deviazione standard. 

### Temporizzare una lista o un `np.array` che contiene 1 milione di lanci di dado 

```python
import numpy as np
```

```python
%timeit rolls_array = np.random.randint(1, 7, 1_000_000)
```

```python
%timeit rolls_list = [np.random.randint(1, 7, 1_000_000)]
```

### Personalizzare %timeit 

```python
%timeit -n3 -r2 rolls_array = np.random.randint(1, 7, 1_000_000)
```

### Altri comandi magici di IPython/Jupyter
IPython/Jupyter hanno dozzine di comandi magici per interagire con il codice e fare svariate operazioni, per esempio:
* **`%load`** per leggere il codice da un file locale o un URL.
* **`%save`** per salvare su file il contenuto del notebook.
* **`%run`** per eseguire uno script .py da dentro Jupyter.
* **`%precision`** per cambiare la precisione in virgola mobile mostrata di default.
* **`%cd`** per cambiare directory senza uscire da Jupyter.
* **`%edit`** per lanciare un editor esterno che permette di modificare notebook pi√π complicati. 
* **`%history`** per vedere tutta la lista di comandi eseguita dall'inizio della sessione.
* **`%jsroot`** per rendere interattivi i grafici di ROOT (lo vedremo dopo).

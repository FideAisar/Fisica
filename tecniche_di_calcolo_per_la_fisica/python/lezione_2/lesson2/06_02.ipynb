## Dizionari autodefiniti
* In Python l'**autodefinizione** Ã¨ l'assegnazione di un'intera collezione di valori (lista, dizionario, ecc.) in una sola istruzione composita.
* Serve a generare collezioni velocemente, spesso **mappando** una collezione a un'altra. 
* L'espressione a sinistra del ciclo `for` specifica un elemento (in questo caso del dizionario) quindi una **coppia chiave-valore _key_`:` _value_**. 
* Esempio: scambiare chiavi a valori in un dizionario con **valori unici**. 

```python
months = {'January': 1, 'February': 2, 'March': 3}
```

```python
months2 = {number: name for name, number in months.items()}
```

```python
months2
```

* In questo altro esempio di autodefinizione, si calcolano le medie dei voti di vari studenti.

```python
grades = {'Sue': [98, 87, 94], 'Bob': [84, 95, 91]}
```

```python
grades2 = {k: sum(v) / len(v) for k, v in grades.items()}
```

```python
grades2
```

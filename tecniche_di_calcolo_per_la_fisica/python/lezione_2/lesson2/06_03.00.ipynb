# Insieme
* Un insieme è una collezione non ordinata di **valori unici**. 
* Può contenere **solo oggetti immutabili**, come stringhe, `int`, `float` e tuple (i cui elementi sono immutabili). 
* Gli insiemi non possono essere indicizzati o sezionati.

### Creare un insieme: parentesi graffe
* Nel crearli viene eseguita automaticamente l'**eliminazione dei duplicati**.

```python
colors = {'red', 'orange', 'yellow', 'green', 'red', 'blue'}
```

* Gli insiemi sono **senza ordine**&mdash;non supportano codice dipendente dall'ordinamento. 

```python
colors
```

### Determinare la grandezza di un insieme

```python
len(colors)
```

### Controllare se l'insieme contiene un elemento

```python
'red' in colors
```

```python
'purple' in colors
```

```python
'purple' not in colors
```

### Iterare su un insieme
* L'ordine di iterazione non ha significato.

```python
for color in colors:
    print(color.upper(), end=' ')
    
```

### Creare un insieme con la funzione `set`

```python
numbers = list(range(10)) + list(range(5))
```

```python
numbers
```

```python
set(numbers)
```

* Per creare un insieme vuoto usare **`set()`**, perché **`{}` rappresenta invece un dizionario vuoto** (come vedremo dopo).
* Python mostra un insieme vuoto come `set()` per evitare confusione con il dizionario vuoto (`{}`).

```python
set()
```

### Operazioni sugli insiemi
* Le più comuni operazioni algebriche sugli insiemi corrispondono a metodi dell'insieme Python, come ad esempio `union`, `intersection`, `difference`, `symmetric_difference` ecc.

# Dizionari
* Un **dizionario** è una collezione non ordinata di **coppie chiave-valore** che mappa dei valori a delle chiavi immutabili, così come un vocabolario mappa parole e definizioni tra loro. 

| Keys | Key type | Values | Value type
| :-------- | :-------- | :-------- | :--------
| Country names | `str` | Internet country codes | `str` 
| Decimal numbers | `int` | Roman numerals | `str` 
| States | `str` | Agricultural products | list of `str` 
| Hospital patients | `str`  | Vital signs | tuple of `int`s and `float`s 
| Baseball players | `str`  | Batting averages | `float` 
| Metric measurements | `str`  | Abbreviations | `str` 
| Inventory codes | `str`  | Quantity in stock | `int` 

* Le chiavi devono essere _immutabili_ e _uniche_.
* Più di una chiave può invece avere lo stesso valore. 

### Creare un dizionario
* Si crea un dizionario includendo tra parentesi graffe, `{}`, una lista separata da virgole di coppie chiave-valore nella forma _key_ : _value_. 
* Si può creare un dizionario vuoto con `{}` e poi aggiungere e/o ridefinire le coppie che lo compongono con istruzioni di assegnazione del tipo `dict[key] = value`

### Operazioni di base

```python
roman_numerals = {'I': 1, 'II': 2, 'III': 3, 'V': 5, 'X': 100}
```

```python
roman_numerals
```

```python
roman_numerals['V']
```

```python
roman_numerals['X'] = 10
```

```python
roman_numerals
```

```python
roman_numerals['L'] = 50
```

```python
roman_numerals
```

```python
del roman_numerals['III']
```

```python
roman_numerals
```

* Il metodo **get** restituisce il valore dell'argomento corrispondente, o `None` se la chiave non è presente.
* **get** con un secondo argomento restituisce quest'ultimo se la chiave non è presente.

```python
roman_numerals.get('III')
```

```python
roman_numerals.get('III', 'III not in dictionary')
```

### Verificare se una chiave è presente nel dizionario

```python
'V' in roman_numerals
```

```python
'III' not in roman_numerals
```

### View di dizionari
* I metodi `items`, `keys` e `values` restituiscono una "view" dei dati del dizionario.
* In generale l'iterazione su una "view" si usa per vedere i contenuti di un oggetto—senza creare una copia dei dati ivi contenuti.

```python
months = {'January': 1, 'February': 2, 'March': 3}
```

```python
for month_name in months.keys():
    print(month_name, end='  ')
```

```python
for month_number in sorted(months.values()):
     print(month_number, end='  ')
```

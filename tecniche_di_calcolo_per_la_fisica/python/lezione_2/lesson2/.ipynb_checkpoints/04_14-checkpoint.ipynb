# La funzione `import`

* `import` rende disponibili al programma funzionalità di altri programmi pacchetti.
* Si può importare un intero programma/classe Python direttamente presente su `$PYTHONPATH`
* Si può importare un intero programma/classe Python di un sotto-pacchetto usando il punto
* Si può importare un solo metodo/classe per volta usando la sintassi `from ... import`

```python
import sys
sys.copyright      ## Note the module.function syntax if using just "import"
```

```python
import matplotlib.pyplot
```

```python
from math import cos
cos(0.)            ## Note: Just call function, not module.function (i.e. just cos, NOT math.cos) if using from ... import
```

### Importare identificatori multipli
* Usare `from ... import` con una lista di identificatori per importarne più di uno

```python
from math import ceil, floor
```

```python
ceil(10.3)
```

```python
floor(10.7)
```

### Import con asterisco
* Si possono importare **tutti** gli identificatori allo stesso momento con **`import *`**.
* Sconsigliato, può portare a errori imprevisti. 

```python
e = 'hello'
```

```python
from math import *
```

```python
e
```

### Legare abbreviazioni ai nomi dei moduli importati  
* Nell'istruzione `import`, la parola **`as`** permette di definire un nome abbreviato per riferirsi al modulo quando vengono chiamati i suoi identificatori. 


```python
import statistics as stats
```

```python
grades = [85, 93, 45, 87, 93]
```

```python
stats.mean(grades)
```

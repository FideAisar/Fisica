# Istruzione `if`
* All'esame di maturità il voto minimo per il superamento è 60. 

> If student’s grade is greater than or equal to 60  
> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Display 'Passed'

* Se la condizione è vera, viene mostrato 'Passed' e si passa poi alle istruzioni seguenti del codice.
* Se la condizione è falsa, non viene mostrato nulla e si passa poi alle istruzioni seguenti del codice.
* L'**indentazione** (rientro di linea) indica quali istruzioni devono essere eseguite solo se la condizione è vera. 

```python
grade = 85
```

```python
if grade >= 60:
    print('Passed')
```

### Indentazione
* La parte indentata delle istruzioni si chiama **suite** (a volte **blocco** in italiano).

```python
if grade >= 60:
print('Passed')  # statement is not indented properly
```

* Attenzione a rientrare sempre dello stesso numero di caratteri.

```python
if grade >= 60:
    print('Passed')
  print('Good job!)
```

### Espressioni abbreviate per `True` o `False`

```python
if 1:
    print('Nonzero values are true, so this will print')
```

```python
if 0:
    print('Zero is false, so this will not print')
```

### Attenzione all'operatore di confronto `==` e quello di assegnazione `=` 
* Usare `==` invece di `=` in un'assegnazione e il viceversa in un confronto può causare problemi difficili da trovare. 
* `grade = 85` (assegna alla variabile il valore 85)
* MA
* `if grade == 85 : ` (controlla se il valore è 85)

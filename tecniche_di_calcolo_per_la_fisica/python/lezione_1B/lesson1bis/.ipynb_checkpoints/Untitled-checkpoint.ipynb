```python
my_ects = { 
    'MAT' : 0,
    'FIS' : 0,
    'CHIM': 0,
    'LIN' : 0,
    'INF' : 0
}
avg_grades = 0.
iline, number_cumlaude = 0, 0
```

```python
with open('./transcripts.txt') as transc :
    lines = transc.readlines()
    for line in lines :
        # print (line.strip())
        if (line.find('SSD') == 0) :
            ssd = line[line.find(' ')+1:line.find('/')]
            nline = lines[iline+1]    # next line
            cfu = int(nline[nline.find(' ')+1:])
            nnline = lines[iline+2]   # next-to--next line
            grades = nnline[nnline.find(' ')+1:nnline.find('/')]
            if (nnline.find('L') > 0) :
                grades = 30;    number_cumlaude += 1
            else if (nnline.find('Approvato') > 0) :
                grades = 0;     cfu = 0;
            if ssd in my_ects[ssd] += cfu
            avg_grades += cfu*grades
        iline += 1        
```

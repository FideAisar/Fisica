# 3.8 Istruzione `for` 
* Ripetere un'azione o una serie di azioni su una **sequenza** di oggetti.

```python
for character in 'Programming':
    print(character, end='  ')
```

* Entrando nel ciclo `for`, Python assegna la 'P' di 'Programming' alla **variabile bersaglio** tra le parole chiave `for` e `in`.
* Dop l'esecuzione della suite, Python assegna al bersaglio il prossimo oggetto nella sequenza (cioè, la '`r`' di '`Programming`'), ed esegure di nuovo la suite. Continua fino a che la lista di oggetti non è finita.

## Iterabili e liste
* La sequenza a destra dell'istruzione `for` deve essere un **iterabile**, cioè una sequenza da cui Python sa prendere nell'ordine un oggetto alla volta. 
* La più semplice collezione ordinata (quindi iterabile) è la **lista**, cioè una collezione di oggetti separati da virgola e racchiusi in parentesi quadre (`[` e `]`). 

```python
total = 0
```

```python
for number in [2, -3, 0, 17, 9]:
    total = total + number
```

```python
total
```

## La funzione `range` 
* Crea un iterabile costituito da interi consecutivi che partono da 0 fino al valore dell'argomento **escluso**.

```python
for counter in range(10):
    print(counter, end=' ')
```

### Function `print`’s `sep` Keyword Argument 
* Keyword argument `sep` (short for separator) specifies the string that appears between the items that print displays. 
* A space character by default. 
* To remove the spaces, use an empty string with no characters between its quotes.

```python
print(10, 20, 30, sep=', ')
```

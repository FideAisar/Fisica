# Troncamento di liste e stringhe
* Si possono **troncare** sequenze (es. liste o stringhe, che sono liste di caratteri) per ottenere nuove sequenze che contengono _sottoinsiemi_ degli elementi originali. 
* Le operazioni di troncamento che _non_ modificano una sequenza funzionano allo stesso modo per liste, tuple e stringhe.

### Troncamento con indice iniziale e finale

```python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]
```

```python
numbers[2:6]
```

### Troncamento con solo l'indice finale o iniziale
* Nel primo caso si assume come indice iniziale `0` 

```python
numbers[:6]
```

```python
numbers[0:6]
```

* Nel secondo caso si assume come indice finale la lunghezza della sequenza.

```python
numbers[6:]
```

```python
numbers[6:len(numbers)]
```

### Troncamento con passo

```python
numbers[::2]
```

### Troncamento con passo o indice negativo

```python
numbers[::-1]
```

```python
numbers[-1:-9:-1]
```

###  APPROFONDIMENTO: Modifica di liste (NON di stringhe o tuple) usando il troncamento

```python
numbers[0:3] = ['two', 'three', 'five']
```

```python
numbers
```

```python
numbers[0:3] = []
```

```python
numbers
```

```python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]
```

```python
numbers[::2] = [100, 100, 100, 100]
```

```python
numbers
```

```python
id(numbers)
```

```python
numbers[:] = []
```

```python
numbers
```

```python
id(numbers)
```

* Deleting `numbers`â€™ contents is different from assigning `numbers` a _new_ empty list `[]`. 
* Identities are different, so they represent separate objects in memory.

```python
numbers = []
```

```python
numbers
```

```python
id(numbers)
```

* When you assign a new object to a variable, the original object will be **garbage collected** if no other variables refer to it.

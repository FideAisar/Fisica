### Esercizio 1
__Fate un unico plot delle due funzioni $f_1(x) = \sin^2(x)/(x^2 +1)$ e $f_2(x) = \sin^2(2\,x -1)/(3\,x^3 +1)$ in 
$\,0\,\le x \le \,5$. Inserite il nome degli assi e le legende.__

````python
import matplotlib.pyplot as plt
import numpy as np

# x,y
x = np.arange(0,6,0.01)

#plots
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x,(np.sin(x)**2)/(x**2 +1),label="$f_1(x) = \sin^2(x)/(x^2 +1)$")
ax.plot(x,(np.sin(2*x - 1)**2)/(3*x**2 +1),label="$f_2(x) = \sin^2(2\,x -1)/(3\,x^3 +1))$")
ax.grid()


# etichette
ax.legend()
ax.set_ylabel('$f(x)$')
ax.set_xlabel('$x$')
ax.set_title('')
````

### Esercizio 2
__La posizione, al tempo t, di un corpo che si muova sotto effetto della gravit terrestre  data da:__
$$y(t) = y_0 + v_{y0} t  \frac{1}{2}\, g\, t^2,\qquad x(t) = x_0 +v_{x0} t$$ 
__dove $g = 9.81 m/s^2$ e $x_0,\,y_0,\,v_{x0},\, v_{y0}$ sono le posizioni e velocit a $t = 0$. 
Supponiamo $y_0 = 100\, m,\,x_0 = 0,\, v_{y0} = 0, v_{x0} = 5\, m/s$. Disegnate la traiettoria $y = y(x)$ del corpo per $y > 0$.
Suggerimento: calcolate come prima cosa il tempo a cui il corpo arriva a $y = 0$.__

````python
x0=0
y0=100
vy0=0
vx0=5
g=9.8

# in t = 4.52, y = 0
t = np.arange(0,4.52,0.1)


y = y0 + vy0*t - (1/2)*g*(t**2)
x = x0 +vx0*t


    
#plots   
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x,y,label="$y(x) = 100   \frac{1}{2}\, g\, t^2,\qquad x(t) = x_0 +v_{x0}$")
ax.grid()


# etichette
ax.legend()
ax.set_ylabel('$y$')
ax.set_xlabel('$x$')
ax.set_title('Traiettoria')
````

### Esercizio 3
__Producete un campione casuale di 100 eventi estratti da una distribuzione normale, con valor medio $\mu = 1.7.$ e deviazione standard $\sigma = 0.4$. Calcolate il valor medio e la deviazione standard del campione. Fate l'istogramma delle frequenze del campione.__

````python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng

#data
rng = default_rng()
data = rng.normal(size=100,scale=0.4,loc=1.7)

media = np.mean(data)
dev = np.std(data)

#plots  
fig, ax = plt.subplots(figsize=(8, 8))
nevent, bins, patches = ax.hist(data, bins=30)


# etichette
ax.plot(np.ones(2)*media,[0,nevent.max()+1],label=f"media = {round(media,2)}")
ax.plot(np.ones(2)*media-dev,[0,nevent.max()+1],label=f"media - $\sigma$ ($\sigma$ = {round(dev,2)})")
ax.plot(np.ones(2)*media+dev,[0,nevent.max()+1],label=f"media + $\sigma$ ($\sigma$ = {round(dev,2)})")

ax.set_ylabel('counts')
ax.set_xlabel('variable')
ax.set_title('')
ax.legend()
````

### Esercizio 4
__Costruite una funzione `f(n)` che generi due array, `x` e `y`, ciascuno di `n `numeri casuali distribuiti uniformemente nell'intervallo fra zero e due. `f` deve ordinare i due array in ordine crescente e fare il grafico usando i due array come coordinate dei punti da plottare.__

````python
def f(n):
    x = np.sort(rng.uniform(0,2,n))
    y = np.sort(rng.uniform(0,2,n))


    #plots  
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.plot(x,y,label="prova")

    # etichette
    ax.set_ylabel('y')
    ax.set_xlabel('x')
    return
f(80)
````

### Esercizio 1
__Trovate le coordinate  x e y dei massimi e dei minimi della curva $\,\,y(x) = \frac{1}{2} \cosh(x) + x^3 -10\, x +5$.__

````python
import numpy as np
from scipy.optimize import fmin
import matplotlib.pyplot as plt

def f(x):
    return (1/2)*np.cosh(x) + x**3 - 10*x + 5
def g(x):
    return -f(x)

x = np.arange(-10, 5, 0.1)


#plot
fig, ax = plt.subplots()
ax.plot(x,f(x))
ax.set_ylim(-100,100)
ax.set_xlabel('x')
ax.grid(visible=True,which='both')

#min/max
min1 = fmin(f, -6,disp=False)
min2 = fmin(f, 2,disp=False)
max1 = fmin(g, -2,disp=False)

print(f"Minimo 1:({min1[0]},{f(min1[0])})")
print(f"Minimo 2:({min2[0]},{f(min2[0])})")
print(f"Massimo:({max1[0]},{f(max1)[0]})")
````

### Esercizio 2
__Trovate le intersezioni della curva $\,\,y(x) = \sinh(x) + x^3 -10\, x +5\,\,$ con l'asse x.__

````python
from scipy.optimize import fsolve

def f(x):
    return np.sinh(x) + x**3 - 10*x + 5

# plot function 
x = np.arange(-10, 10, 0.1)
y = f(x)

fig, ax = plt.subplots(figsize=(6,4))
ax.set_xlabel('x')
ax.grid(visible=True,which='both')
ax.set_ylim(-100,100)

ax.plot(x, f(x))

#zeri
x1 = -2.5
x2 = 0
x3 = 2.5
z1 = fsolve(f, x1)   
z2 = fsolve(f, x2)   
z3 = fsolve(f, x3)   

print(f"Zeri della funzione: {z1[0]},{z2[0]},{z3[0]}")
````

### Esercizio 3
__Disegnate la curva $\,\,y(x) = x^2\cos(x/2) + 2\,x  +1\,\,$ e calcolate l'area compresa fra  la curva e l'asse $x$ nel rettangolo $y \ge 0$; $\, -5 \le x \le 5$.__

````python
from scipy.integrate import quad 

def f(x):
    return (x**2)*np.cos(x/2) + 2*x + 1 

# plot function 
x = np.arange(-10, 10, 0.1)

fig, ax = plt.subplots(figsize=(6,4))
ax.set_xlabel('x')
ax.grid(visible=True,which='both')

ax.plot(x, f(x))
ax.plot(x,np.zeros(200))

# zeri
x1 = -1
x2 = 4
z1 = fsolve(f, x1)   
z2 = fsolve(f, x2)   

#l'area tra 0 e 5 con y>0  l'area tra i due zeri
print(quad(f, z1, z2))
````

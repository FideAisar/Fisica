### Esercizio 1
__Fate il plot delle due funzioni $f_1(x) = (x^2-x+2)/(x^4+2)$ in $\,0\,\le x \le \,4$ e $f_2(x) = \cos(x+1)/(x^2+ 1)$ in $\,-2\,\le x \le \,2$ nello stesso grafico. Inserire un titolo per il plot, i titoli degli assi e le legende.__

````python
import matplotlib.pyplot as plt
import numpy as np

# x,y
x = np.arange(0,6,0.01)

#plots
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x,(x**2 - x + 2)/(x**4 + 2),label="$f_1(x) = (x^2-x+2)/(x^4+2)$")
ax.plot(x,np.cos(x+1)/(x**2 + 1),label="$f_2(x) = \cos(x+1)/(x^2+ 1)$")
ax.grid()


# etichette
ax.legend()
ax.set_ylabel('$f(x)$')
ax.set_xlabel('$x$')
ax.set_title('Titolo')
````

### Esercizio 2
__Generate le coordinate x e y di 100 punti casuali uniformemente distribuiti nel rettangolo di vertici (0,0) (0,5) (8,0) (8,5).__<BR>
__Fate il plot dei punti generati.__<BR>
__Calcolate il valor medio della distanza dei punti dall'origine.__<BR>


````python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng

rng = default_rng()

x = rng.uniform(0,8,100)
y = rng.uniform(0,5,100)

#plots   
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(x,y)
ax.grid()


# etichette
ax.set_ylabel('$y$')
ax.set_xlabel('$x$')
ax.set_title('100 punti casuali uniformemente distribuiti')

dist = np.sqrt(x**2 + y**2)

print("la distanza media di un punto dall'origine Ã¨ di",np.mean(dist))
````

### Esercizio 3
__Fate il plot delle funzioni $f_n(x) = (\sin(x^2-x+2))^n/(x^4+2)$ con $n$ = 1,2,3,4 in quattro plot disposti su due righe.Inserire i titoli degli assi e le legende. Salvate il plot su disco in formato pdf.__

````python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng

#data
x = np.arange(-5,5,0.01)
y = np.array([(np.sin(x**2 - x + 2)**n) / (x**4 + 2) for n in range(1,5)])



#plots   
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

axes = [ax1,ax2,ax3,ax4]

for i in axes:
    i.plot(x,y[axes.index(i)])
    i.set_ylabel('$y$')
    i.set_xlabel('$x$')
    i.set_title(f'$f_n(x) = sin(x^2-x+2)^{axes.index(i) + 1}/(x^4+2)$')
    

````

### Esercizio 4
__Producete due campion1 casual1 di 500 eventi ciascuno, estratti da due distribuzioni normali, con valor medio $\mu_1 = 1.7$ e deviazione standard $\sigma_1 = 0.4$ e con valor medio $\mu_2 = 2.1$ e deviazione standard $\sigma_2 = 0.7$ rispettivamente. Fate la somma elemento per elemento dei due campioni. Calcolate il valor medio e la deviazione standard del risultato. Fate l'istogramma delle frequenze del risultato.__

````python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng

#data
rng = default_rng()
data1 = rng.normal(size=500,scale=0.4,loc=1.7)
data2 = rng.normal(size=500,scale=0.7,loc=2.1)

somma = data1 + data2

media = np.mean(somma)
dev = np.std(somma)

#plots  
fig, ax = plt.subplots(figsize=(8, 8))
nevent, bins, patches = ax.hist(somma, bins=30)


# etichette
ax.plot(np.ones(2)*media,[0,nevent.max()+1],label=f"media = {round(media,2)}")
ax.plot(np.ones(2)*media-dev,[0,nevent.max()+1],label=f"media - $\sigma$ ($\sigma$ = {round(dev,2)})")
ax.plot(np.ones(2)*media+dev,[0,nevent.max()+1],label=f"media + $\sigma$ ($\sigma$ = {round(dev,2)})")

ax.set_ylabel('counts')
ax.set_xlabel('variable')
ax.set_title('')
ax.legend()
````

### Esercizio 1
__La funzione degli errori `erf` è definita dall'integrale:__

$$ erf(x) = \frac{2}{\sqrt{\pi}}\, \int_0^x e^{-t^2} dt.$$

__Disegnate la funzione $erf(x)$, facendo gli integrali necesari, in $\, 0 \le x \le 2$.__<BR>
__Nota: ciascun estremo di integrazione nella chiamata a `quad` deve essere un singolo numero reale. Non può essere un array.
In questo caso è quindi necessario avere un ciclo `for`.__

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy

t = np.arange(0,2,0.01)
y = []

def f(t):
    return np.e**(-t**2)


# plot
fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlabel('x')
ax.grid(visible=True,which='both')

ax.plot(t, f(t))
````

```
    pippo = np.array([0.,1.,2.5,2.8,3.0,3.5,3.9,4.1,5.7,6.0,6.8,7.7,8.2,9.5,10.0])
```
__Costruite l'istogramma delle frequenze di `dX` se `X` è `pippo`.__

### Esercizio 3
__Disegnate la curva $\,\,y(x) = - x^2 +\, x\, e^{-x^2} +2 \,\,$ e calcolate l'area compresa fra la curva  e l'asse $\,x\,$ nel semipiano $\,y \ge 0\,$__ 

````python
def f(x):
    return -x**2+x*np.e**(-x**2)+2

x = np.arange(-5,5,0.01)
y = f(x) 
fig, ax = plt.subplots(figsize=(7,7))

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()
ax.set_xlim(-5,5)
ax.set_ylim(0,5)

ax.plot(x, y)
````

````python
from scipy.integrate import quad
from scipy.optimize import fsolve
xstart1 = -2
x1 = fsolve(f, xstart1)

xstart2 = 1
x2 = fsolve(f, xstart2)

res = quad(f,x1,x2)
print(res)
````
